```python
# Import the library for encoding
from mlxtend.preprocessing import TransactionEncoder

# Instantiate transaction encoder
encoder = TransactionEncoder().fit(transactions)
```

```python
# One-hot encode itemsets by applying fit and transform
onehot = encoder.transform(transactions)

# Convert one-hot encoded data to DataFrame
onehot = pd.DataFrame(onehot, columns = encoder.columns_)
print(onehot)
```

## Apriori

```python
# Import Apriori algorithm
from mlxtend.frequent_patterns import apriori
```

```python
# Compute frequent itemsets
#frequent_itemsets1 = apriori(onehot, min_support = 0.0005,max_len = 4, use_colnames = True)
frequent_itemsets = apriori(onehot, min_support = 0.005,max_len = 5, use_colnames = True)

# Print number of itemsets
print(len(frequent_itemsets))
```

```python
import sys
```

```python
sys.getsizeof(frequent_itemsets) 
```

```python
print(frequent_itemsets.head())
```

```python
# Import Apriori algorithm
from mlxtend.frequent_patterns import apriori, association_rules

# Compute association rules
Rules = association_rules(frequent_itemsets,
                          metric = "support",
                          min_threshold = 0.005)
```

```python
# Print association rules
Rules
```

```python
filtered_rules = Rules[(Rules['antecedent support'] > 0.01) & 
                       (Rules['support'] > 0.009) & 
                       (Rules['confidence'] > 0.5) & 
                       (Rules['lift'] > 1.00)]
```

```python
filtered_rules
```

```python
type(filtered_rules)
```

```python
final_rules = filtered_rules.sort_values('confidence',ascending=False)
```

```python
def csv_output(final_dataframe,part_dict,df_result):
    
    
    for index_number in range(len(final_dataframe)):
        x = final_dataframe.iloc[index_number]['antecedents']
        y = final_dataframe.iloc[index_number]['consequents']
        for i in x:
             part_name = part_dict[i][0]
             temp_a = str(i)+" "+str(part_name)
             df_result.iloc[index_number]['antecedents'] = temp_a
        for i in y:
             part_name = part_dict[i][0]
             temp_c = str(i)+" "+str(part_name)
             df_result.iloc[index_number]['consequents'] = temp_c
    df_result['support'] = final_dataframe['support']
    df_result['confidence'] = final_dataframe['confidence']
    df_result['lift'] = final_dataframe['lift']
    return df_result
    

    
```
